```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import random as rm
import math
```

# Problema 1

Para la resolución de este problema primero generé una variable aleatoria con distribución uniforme que va desde 0 a 1 (p); luego con ella generé una nueva variable aleatoria (z) que va desde -pi/2 hasta pi/2.

De esta manera, generé 10000 números al azar con distribución de Cauchy aplicando la tangente a la variable aleatoria "z", y los guardé en una lista.

```python
var_alea_cauchy = []
N = 10000
while len(var_alea_cauchy)<= N:
    p = rm.uniform(0,1)
    z = math.pi * p - math.pi/2
    var_alea_cauchy.append(math.tan(z))


# Para graficar el histograma.
alt, anch, _ = plt.hist (var_alea_cauchy, bins = 30, density = True, edgecolor = "black", range = (-2*math.pi,2*math.pi))
plt.xlabel ("Valor de la variable")
plt.ylabel ("Número de veces que se obtuvo")
plt.title ("Histograma de una variable \n aleatoria con distribución de Cauchy")


# Para graficar la Distribución de Cauchy.
x = np.arange (min(anch),max(anch),0.00001)
y = stats.cauchy.pdf(x)


# Para calcular error del histograma.
centrobin = 0.5*(anch[1:]+anch[:-1]) 
# El array "anch" tiene 1 elemento mas que el array "alt". Toma en cuenta los extremos. Tomo el centro del bin.
sigma_histograma = []
for i in range (len(alt)):
    sigma_histograma.append(math.sqrt(N*alt[i]*(1-alt[i]))/N) #REVISAAAAAAAAAAAAAAR

plt.errorbar (centrobin,alt,yerr = sigma_histograma, fmt = " .r") 
plt.plot (x,y)


plt.show()
```

Dibujo el histograma en dicho intervalo, desde -1.75pi hasta 1.75pi, para que su normalización corresponda lo mas posible con la pdf de Cauchy. La función de stats que utilizo para el ajuste, esta normalizada integrando desde -00 a +00, sin embargo los datos que yo estoy viendo no van desde -00 a +00; lo cual corrigo agrandando el intervalo que ploteo en el histograma.
Al hacer crecer el intervalo y tomar mas datos la normalización de este se corresponde con la normalización de la función ya que se pueden despreciar la contribución de los valores mas grandes.

Otra forma de resolverlo implica integrar la función de Cauchy, desde "a" hasta "b" y multiplicar por ese factor a mi función
de stats , quedando de esta manera escalada con el histograma que va desde "a" hasta "b".

```python
# Montecarlo.
a = -2* math.pi
b =  2* math.pi
dominio = [a,b]
fm = 1/math.pi

N = 10000
num_gene = []
while len(num_gene) < N:
    y = rm.uniform(0,1)
    z = rm.uniform(0,1)
    
    u = dominio [0] + (dominio[1]-dominio[0])*y
    v = fm * z
    
    if v <= stats.cauchy.pdf(u):
        num_gene.append(u)

# Para graficar el histograma.
alt, anch, _ = plt.hist (num_gene, bins = 30,edgecolor = "black", density = True, range = (-2*math.pi,2*math.pi))
plt.xlabel ("Valor de la variable")
plt.ylabel ("Número de veces que se obtuvo")
plt.title ("Histograma de una variable aleatoria \n con distribución de Cauchy - Montecarlo")


# Para graficar la Distribución de Cauchy.
x = np.linspace (min(anch),max(anch),500)
y = stats.cauchy.pdf(x)

plt.plot (x,y)

plt.show()
```

1. Se puede ver la diferencia entre utilizar el método montecarlo y el método analítico, ya que analizando el mismo intervalo en el método analítico obtenemos un ajuste mucho mas preciso.

# Problema 2

1ro. Ajusté linealmente los datos obteniendo la ordenada al origen (a1) y la pendiente (a2).

2do. Utilicé la matriz de covarianza de a1 y a2 para calcular los errores en la estimación de a1 y a2.

3ro. Plotee la recta con el valor hallado para la pendiente y la ordenada.

```python
## PUNTO 1

## ------------ Datos iniciales ------------ ##
x = np.linspace(2,3,11)
y = np.asarray([2.78,3.29,3.29,3.33,3.23,3.69,3.46,3.87,3.62,3.40,3.99])
sigma_y = 0.3
a2,a1,_,_,_ = stats.linregress (x,y)


## ------------ Cálculo auxiliares para la matriz de covarianza de los parámetros ------------ ##
suma_cuadrados_x = 0
for i in range (len(x)):
    aux1 = x[i]**2
    suma_cuadrados_x = suma_cuadrados_x + aux1

suma_x = sum (x)
N = len(x)
delta = N*suma_cuadrados_x - suma_x **2

## ------------ Cálculo de la matriz de covarianza de los parámetros ------------ ##
sigma_a1 = math.sqrt (sigma_y**2/delta * suma_cuadrados_x)
sigma_a2 = math.sqrt (sigma_y**2/delta * N)


## ------------ Recta obtenida a partir de cuadrados mínimos ------------ ##
fx = lambda x: a1 + a2 * x
x_recta = np.linspace(0,5,25)
y_recta = []
for i in range (len(x_recta)):
    y_recta.append(fx(x_recta[i]))



plt.plot (x_recta,y_recta)
plt.errorbar (x,y, yerr = sigma_y, fmt='.k')
plt.grid()
plt.show()

print ("a1:",a1,"Error a1:",sigma_a1,"a2:",a2,"Error a2:",sigma_a2)
```

1ro. Dibujé la recta del inciso anterior mas los datos. [Negro]

2do. Usando dicha recta calculé los valores de Y_a para valores de X_a distintos de los datos.

3ro. Grafiqué dichos Y_a con su error (teniendo en cuenta el factor de correlación entre los parámetros a1 y a2). [Verde]

4to. Grafiqué los valores de Y_a (solo por cuestiones de visualización calculé Y_a para X_a + dx) con su error pero si el factor de correlación. [Rojo]

5to. Aclaración: Por cuestiones de visualización los Y_a no se tomaron para valores de X_a comprendidos entre el mínimo y máximo de los datos.

```python
## PUNTO 2 y 3.

## ------------ Datos iniciales ------------ ##
x = np.linspace(2,3,11)
y = np.asarray([2.78,3.29,3.29,3.33,3.23,3.69,3.46,3.87,3.62,3.40,3.99])
sigma_y = 0.3
a2,a1,_,_,_ = stats.linregress (x,y)


## ------------ Cálculo auxiliares para la matriz de covarianza de los parámetros ------------ ##
suma_cuadrados_x = 0
for i in range (len(x)):
    aux1 = x[i]**2
    suma_cuadrados_x = suma_cuadrados_x + aux1

suma_x = sum (x)
N = len(x)
delta = N*suma_cuadrados_x - suma_x **2

## ------------ Cálculo de la matriz de covarianza de los parámetros ------------ ##
sigma_a1 = math.sqrt (sigma_y**2/delta * suma_cuadrados_x)
sigma_a2 = math.sqrt (sigma_y**2/delta * N)
cov_a1_a2 = - suma_x * sigma_y**2 / delta


## ------------ Recta obtenida a partir de cuadrados mínimos ------------ ##
fx = lambda x: a1 + a2 * x
x_recta = np.linspace(0,5,50)
y_recta = []
for i in range (len(x_recta)):
    y_recta.append(fx(x_recta[i]))

plt.plot (x_recta,y_recta)
plt.errorbar (x,y, yerr = sigma_y, fmt='.k')
## ------------ Hasta aca para graficar la recta del item anterior ------------ ##

## ------------ Predicción y_a(x_a) con correlación entre los parámetros ------------ ##
aux2 = np.linspace(0,5,50)
x_a = []
y_a = []
sigma_y_a = []
for i in range (len(aux2)):
    if aux2[i] < min (x) or aux2[i] > max (x):
        y_a.append(fx(aux2[i]))
        x_a.append(aux2[i])
        sigma_y_a.append(math.sqrt(sigma_a1 **2 + aux2[i]**2 * (sigma_a2)**2 + 2 * aux2[i] * cov_a1_a2))

plt.errorbar(x_a,y_a,yerr = sigma_y_a,fmt='.g')

indice_minimo_sigma = sigma_y_a.index(min(sigma_y_a))
x_menor_sigmaY = x_a[indice_minimo_sigma] # Valor de x para el cual obtengo el menor valor de sigma_y_a.
print ("Xa cuyo valor de Ya tiene mínimo error:",x_menor_sigmaY)

## ------------ Predicción y_a(x_a) con correlación nula entre los parámetros ------------ ##
aux2 = np.linspace(0,5,20)
dx = 0.1
x_a_sc = []
y_a_sc = []
sigma_y_a_sc = []
for i in range (len(aux2)):
    if aux2[i] < min (x) or aux2[i] > max (x):
        y_a_sc.append(fx(aux2[i]+dx))
        x_a_sc.append(aux2[i]+dx)
        sigma_y_a_sc.append(math.sqrt(sigma_a1 **2 + aux2[i]**2 * (sigma_a2)**2))

plt.errorbar(x_a_sc,y_a_sc,yerr = sigma_y_a_sc,fmt='*r')

plt.grid()
plt.show()
```

1. Con correlación se observa que a mayor distancia de los datos, las predicciones tienen mas error; tanto para valores de Xa mas grandes, o mas chicos, que los de nuestros datos. Que el error aumente para valores de Xa mas alejados de los datos se debe a que la covarianza entre "a1" y "a2" es negativa, pero su contribución al sigma_y_a es cada vez menor para valores de Xa mas alejados ya que el término (positivo) que acompaña a la varianza de "a2" es Xa al cuadrado mientrás que el que acompaña a la covarianza entre "a1" y "a2" es Xa sólo. De hecho como puede observarse, el valor de Xa que minimiza el error de Y_a es el que se encuentra mas próximo a los datos.


2. Sin correlación (las líneas rojas) a igual distancia de los datos, el error para valores mas grandes de Xa es mayor que para valores mas chicos. Pudiéndose inferir que la predicción es tanto mas precisa cuanto mas nos alejamos de los datos con valores mas pequeños de Xa.
