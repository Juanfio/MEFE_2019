```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import random as rm
import math
```

```python
var_alea_cauchy = []
while len(var_alea_cauchy)<= 10000:
    p = rm.uniform(0,1)
    z = math.pi * p - math.pi/2
    if -math.pi/(2.30) < z < math.pi/(2.30):
        var_alea_cauchy.append(math.tan(z))



# Para graficar el histograma.
alt, anch, _ = plt.hist (var_alea_cauchy, bins = 20, normed = True, edgecolor = "black") 
plt.xlabel ("Valor de la variable")
plt.ylabel ("Número de veces que se obtuvo")
plt.title ("Histograma de una variable \n aleatoria con distribución de Cauchy")

# Para chequear que este efectivamente normalizado
anch_2 = []
for i in range (len(anch)-1):
    anch_2.append(anch[i+1]-anch[i])

areas = []
for i in range (len(anch_2)):
    areas.append (anch_2[i]*alt[i])

factor_normalizacion = sum (areas)
print (factor_normalizacion)

# Para graficar la Distribución de Cauchy.
x = np.arange (anch[0],anch[len(anch)-1],0.00001)
y = stats.cauchy.pdf(x)
# # y_2 = math.pi * y

plt.plot (x,y)


plt.show()
```

```python
# Montecarlo.
a = 0
b = 2
dominio = [a,b]
fm = 1

num_gene = []
while len(num_gene) < 1000:
    y = rm.uniform(0,1)
    z = rm.uniform(0,1)
    
    u = dominio [0] + (dominio[1]-dominio[0])*y
    v = fm * z
    
    if v <= stats.cauchy.pdf(u):
        num_gene.append(u)

# Para graficar el histograma.
alt, anch, _ = plt.hist (num_gene, edgecolor = "black", normed = True)
plt.xlabel ("Valor de la variable")
plt.ylabel ("Número de veces que se obtuvo")
plt.title ("Histograma de una variable aleatoria \n con distribución de Cauchy - Montecarlo")

# Para chequear que este efectivamente normalizado
anch_2 = []
for i in range (len(anch)-1):
    anch_2.append(anch[i+1]-anch[i])

areas = []
for i in range (len(anch_2)):
    areas.append (anch_2[i]*alt[i])

factor_normalizacion = sum (areas)
print (factor_normalizacion)

# Para graficar la Distribución de Cauchy.
x = np.arange (a,b,0.00001)
y = stats.cauchy.pdf(x)
y_2 = math.pi * y

plt.plot (x,y_2)


plt.show()
```
