```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import random as rm
import math
```

# G3P10 - Monte Carlo.

```python
var_alea_cauchy = []
N = 10000
while len(var_alea_cauchy)<= N:
    p = rm.uniform(0,1)
    z = math.pi * p - math.pi/2
    var_alea_cauchy.append(math.tan(z))


# Para graficar el histograma.
alt, anch, _ = plt.hist (var_alea_cauchy, bins = 20, density = True, edgecolor = "black", range = (-1.75*math.pi,1.75*math.pi))
# JUSTIFICACIÓN: Utilizo ese intervalo, desde -1.75pi hasta 1.75pi, porque sino la función queda fuera de escala. La función
# de stats que utilizo para el ajuste, esta normalizada integrando desde -00 a +00, sin embargo los datos que yo estoy viendo
# no van desde -00 a +00, por lo tanto es logico que no este en escala. Al hacer crecer el intervalo y tomar mas datos
# la normalización de dicho histograma se corresponde con la normalización de la función ya que se pueden despreciar la
# contribución de los valores mas grandes.
# Otra forma de resolverlo implica integrar la función de Cauchy, desde "a" hasta "b" y multiplicar por ese factor a mi función
# de stats , quedando de esta manera escalada con el histograma que va desde "a" hasta "b".
plt.xlabel ("Valor de la variable")
plt.ylabel ("Número de veces que se obtuvo")
plt.title ("Histograma de una variable \n aleatoria con distribución de Cauchy")

# Para chequear que este efectivamente normalizado
anch_2 = []
for i in range (len(anch)-1):
    anch_2.append(anch[i+1]-anch[i])

areas = []
for i in range (len(anch_2)):
    areas.append (anch_2[i]*alt[i])

factor_normalizacion = sum (areas)

# Para graficar la Distribución de Cauchy.
x = np.arange (min(anch),max(anch),0.00001)
y = stats.cauchy.pdf(x)



centrobin = 0.5*(anch[1:]+anch[:-1]) 
# El array "anch" tiene 1 elemento mas que el array "alt". Toma en cuenta los extremos. Tomo el centro del bin.
sigma_histograma = []
for i in range (len(alt)):
    sigma_histograma.append(math.sqrt(N*alt[i]*(1-alt[i]))/N) #REVISAAAAAAAAAAAAAAR

plt.errorbar (centrobin,alt,yerr = sigma_histograma, fmt = " .r") 

plt.plot (x,y)


plt.show()
```

```python
# Montecarlo.
a = -1.75* math.pi/2
b =  1.75* math.pi/2
dominio = [a,b]
fm = 1/math.pi

num_gene = []
while len(num_gene) < 1000:
    y = rm.uniform(0,1)
    z = rm.uniform(0,1)
    
    u = dominio [0] + (dominio[1]-dominio[0])*y
    v = fm * z
    
    if v <= stats.cauchy.pdf(u):
        num_gene.append(u)

# Para graficar el histograma.
alt, anch, _ = plt.hist (num_gene, bins = 20,edgecolor = "black", density = True, range = (-1.75*math.pi,1.75*math.pi))
plt.xlabel ("Valor de la variable")
plt.ylabel ("Número de veces que se obtuvo")
plt.title ("Histograma de una variable aleatoria \n con distribución de Cauchy - Montecarlo")

# Para chequear que este efectivamente normalizado
anch_2 = []
for i in range (len(anch)-1):
    anch_2.append(anch[i+1]-anch[i])

areas = []
for i in range (len(anch_2)):
    areas.append (anch_2[i]*alt[i])

factor_normalizacion = sum (areas)

# Para graficar la Distribución de Cauchy.
x = np.linspace (min(anch),max(anch),500)
y = stats.cauchy.pdf(x)

plt.plot (x,y)

plt.show()
```

1. Se puede ver la diferencia entre utilizar el método montecarlo y el método analítico, ya que analizando el mismo intervalo en el método analítico obtenemos un ajuste mucho mas preciso.
